```python
import pandas as pd

# Create the star data manually (same data as the CSV)
data = {
    "Name": ["Sirius", "Canopus", "Rigil Kentaurus", "Arcturus", "Vega",
             "Capella", "Rigel", "Procyon", "Achernar", "Betelgeuse"],
    "Magnitude": [-1.46, -0.72, -0.27, -0.05, 0.03,
                  0.08, 0.12, 0.38, 0.46, 0.50],
    "Distance_ly": [8.6, 310, 4.4, 37, 25,
                    43, 860, 11.4, 140, 642.5],
    "RA": [101.3, 95.99, 219.9, 213.9, 279.2,
           79.2, 78.6, 114.8, 24.4, 88.8],
    "Dec": [-16.7, -52.7, -60.8, 19.2, 38.8,
            46.0, -8.2, 5.2, -57.2, 7.4]
}

stars = pd.DataFrame(data)

# Show the first few rows
stars.head()
```

```python
# Sort by brightness (lowest magnitude is brightest)
brightest_five = stars.sort_values(by="Magnitude").head(5)

# Show the top 5 brightest stars
brightest_five
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(10, 6))

# Plot the stars
plt.scatter(brightest_five["RA"], brightest_five["Dec"],
            s=200, c='gold', edgecolors='black', label="Bright Stars")

# Add labels
for i, row in brightest_five.iterrows():
    plt.text(row["RA"] + 1, row["Dec"], row["Name"], fontsize=12)

# Title and axes
plt.title("Top 5 Brightest Stars in the Sky", fontsize=16)
plt.xlabel("Right Ascension (RA)")
plt.ylabel("Declination (Dec)")
plt.grid(True)
plt.legend()
plt.show()
```

In this beginner-friendly space project, I explored a dataset of bright stars and identified the five brightest based on their apparent magnitude (the lower the value, the brighter the star appears from Earth).

Using Python and pandas, I sorted the stars and visualized them on a 2D celestial map using Right Ascension (RA) and Declination (Dec).

The stars I found include well-known names like Sirius, Canopus, and Vega. These stars are not only the brightest but also relatively close to Earth in light-years.
